```python
# seaborn and matplotlib combined
# tips dataset also avaiable in defau;t in PP
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
tips
```

```python
# creating bar graph for iris

plt.bar(data['species'], data['petal_length'])
plt.bar(data['species'], data['petal_width'])
plt.show()
```

```python
plt.bar(tips["day"], tips['total_bill'])
plt.show()
```

```python
plt.barh(tips["day"], tips['total_bill'])
plt.show()
```

```python
# Histogram

plt.hist(tips['total_bill'], edgecolor = "black")
plt.show()
```

```python
#Day must be in fixed length

day = tips.groupby('day')['total_bill'].sum().reset_index()

day
```

```python
#Pie chart

x = day['day']

y = day['total_bill']
```

```python
explode = [0,0,0,0.1]

plt.pie(y,labels=x,explode = explode,autopct='%1.1f%%',colors=['red','blue','green','yellow'],
        wedgeprops = {'edgecolor' : 'black'},shadow=True,startangle=90)
plt.show()
```

```python
# scatter plot

plt.scatter(tips['total_bill'], tips['tip'])
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

```python
# create scatter plot with color pallete

# scatter plot

tips['size'].value_counts()
plt.scatter(tips['total_bill'], tips['tip'], s = 100, c = tips['size'])
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
cbar = plt.colorbar()
cbar.set_label("Size")

plt.show()
```
